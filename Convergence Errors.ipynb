# # Convergence Plots
# The following cells generate logâ€“log plots for spatial and temporal convergence.

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample convergence data (replace with your actual results)
h = np.array([0.02, 0.01, 0.005, 0.0025])
E_h = np.array([1.23e-3, 3.10e-4, 7.76e-5, 1.94e-5])

dt = np.array([1.00e-3, 5.00e-4, 2.50e-4, 1.25e-4])
E_dt = np.array([4.56e-4, 1.15e-4, 2.88e-5, 7.20e-6])

# %% [code]
# Spatial convergence plot
plt.figure()
plt.loglog(h, E_h, marker='o')
plt.xlabel('Grid spacing $h$')
plt.ylabel('Error $E_h$')
plt.title('Spatial Convergence (Expected $\\mathcal{O}(h^2)$)')
plt.grid(True, which="both", ls="--")
plt.show()

# %% [code]
# Temporal convergence plot
plt.figure()
plt.loglog(dt, E_dt, marker='o')
plt.xlabel('Time step $\\Delta t$')
plt.ylabel('Error $E_{\\Delta t}$')
plt.title('Temporal Convergence (Expected $\\mathcal{O}(\\Delta t^2)$)')
plt.grid(True, which="both", ls="--")
plt.show()

```
